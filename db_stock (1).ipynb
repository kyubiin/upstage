```python
# 삼성전자의 corp_code 가져오기
corp_code = df_listed[df_listed['corp_name'] == '삼성전자'].iloc[0, 0]

# 배당 현황 데이터 가져오기 (2023년 4분기 기준 예시)
data = dart_fss.api.info.alot_matter(corp_code, '2023', '11011')
df = pd.DataFrame(data['list'])

# 필요한 칼럼 이름을 정의 (se 값에 기반)
required_columns = [
    '주당액면가액(원)',
    '(연결)당기순이익(백만원)',
    '(별도)당기순이익(백만원)',
    '(연결)주당순이익(원)',
    '현금배당금총액(백만원)'
]

# 필요한 데이터만 필터링
filtered_data = df[df['se'].isin(required_columns)].copy()

# 새로운 데이터프레임에 각 컬럼을 추가하고 thstrm 값을 할당
result1 = pd.DataFrame({
    'corp_code': [corp_code],
    'corp_name': ['삼성전자']
})

for col in required_columns:
    value = filtered_data[filtered_data['se'] == col]['thstrm'].values
    result1[col] = value[0] if len(value) > 0 else None  # 값이 존재하면 할당, 아니면 None

# print(result)
# print(filtered_data)

result1
```

```python
import pandas as pd

# 종목 코드와 종목명 리스트
codes = ['00126380', '00164779', '01596425', '00161383', '00369657', '00246417', '00535676', '00160843', '00252135', '01489648', '00118804']
names = ['삼성전자', 'SK하이닉스', 'SK스퀘어', '한미반도체', '리노공업', '이오테크닉스', '테크윙', 'DB하이텍', '주성엔지니어링', '솔브레인', '동진쎄미켐']

# 결과를 저장할 리스트 초기화
all_sales_data = []

# 각 종목 코드에 대해 반복
for corp, name in zip(codes, names):
    # 각 종목의 데이터를 가져오기
    financial_data = dart.finstate(corp, int('2024'), '11013')  # bsns_year를 정수로 변환
    
    # 데이터가 조회되지 않는 경우를 처리
    if financial_data.empty:
        print(f"{name}에 대한 데이터가 없습니다.")
        continue
    
    # 필요한 열만 선택하고 매출액 행 필터링
    financial_data_final = financial_data[['account_nm', 'thstrm_nm', 'thstrm_dt', 'thstrm_amount']]
    sales_data = financial_data_final[financial_data_final['account_nm'] == '매출액']
    
    # 종목 코드와 종목명을 추가
    sales_data['종목코드'] = corp
    sales_data['종목명'] = name
    
    # 결과 리스트에 추가
    all_sales_data.append(sales_data)

# 리스트를 하나의 데이터프레임으로 결합
result= pd.concat(all_sales_data, ignore_index=True)
result = result[['종목코드', '종목명', 'account_nm', 'thstrm_nm', 'thstrm_dt', 'thstrm_amount']]

# 'thstrm_amount' 컬럼을 숫자 형식으로 변환
result['thstrm_amount'] = result['thstrm_amount'].replace(',', '', regex=True).astype(float)

# 종목별로 평균 계산 및 첫 번째 값 가져오기
result_avg = result.groupby(['종목코드', '종목명'], as_index=False).agg({
    'thstrm_amount': 'mean',
    'account_nm': 'first',
    'thstrm_nm': 'first',
    'thstrm_dt': 'first'
})

# 최종 결과 출력
result_avg
```

```python
import sqlite3
import pandas as pd

# 예제 데이터프레임 정의
data = {
    '종목코드': [],
    '종목명': [],
    '주당액면가액(원)': [],
    '(연결)당기순이익(백만원)': [],
    '(별도)당기순이익(백만원)': [],
    '(연결)주당순이익(원)': [],
    '현금배당금총액(백만원)': [],
    '당기금액': [],
    '계정명': [],
    '당기명': [],
    'thstrm_dt': [],
    '개괄': [],
    '투자 의견': [],
    '현재가': [],
    'Close': [],
    '%Change': [],
    'Market cap.': []
}

# 실제 데이터프레임을 여기에 할당하세요
final_data = pd.DataFrame(data)

# SQLite 데이터베이스 연결
conn = sqlite3.connect('info_stock.db')

# 데이터프레임을 SQLite 데이터베이스의 테이블로 저장
final_data.to_sql('stock_info', conn, if_exists='replace', index=False)

# 데이터베이스 연결 종료
conn.close()
print("데이터가 info_stock.db에 성공적으로 저장되었습니다.")
```
